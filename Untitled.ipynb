```python
ca = gpd.read_file(r"counties.geojson")
```

```python
df = pd.read_csv(r'test.csv')
```

```python
df = df[~df['County'].isnull()]
```

```python
df = df.groupby(['County','Plan Group','Time Period: Incurred Year']).sum().reset_index()
```

```python
df['Time Period: Incurred Year'] = df['Time Period: Incurred Year'].astype(int)
```

```python
a18 = df[(df['Plan Group'] == 'Plan A') & (df['Time Period: Incurred Year'] ==2018)]
```

```python
a19 = df[(df['Plan Group'] == 'Plan A') & (df['Time Period: Incurred Year'] ==2019)]
```

```python
b18 = df[(df['Plan Group'] == 'Plan B') & (df['Time Period: Incurred Year'] ==2018)]
```

```python
b19 = df[(df['Plan Group'] == 'Plan B') & (df['Time Period: Incurred Year'] ==2019)]
```

```python
c18 = df[(df['Plan Group'] == 'Plan C') & (df['Time Period: Incurred Year'] ==2018)]
```

```python
c19 = df[(df['Plan Group'] == 'Plan C') & (df['Time Period: Incurred Year'] ==2019)]
```

```python
y = 'a18'
a18 = a18.rename(columns={
'Plan Group':'PG_'+y,
'Time Period: Incurred Year':'date_'+y,
'Member Months':'mm_'+y,
'Dollars':'dollars_'+y,
'ER Visits':'er_'+y,
'Admits':'admits_'+y,
'Days Admit':'dayadmit_'+y})

y = 'a19'
a19 = a19.rename(columns={
'Plan Group':'PG_'+y,
'Time Period: Incurred Year':'date_'+y,
'Member Months':'mm_'+y,
'Dollars':'dollars_'+y,
'ER Visits':'er_'+y,
'Admits':'admits_'+y,
'Days Admit':'dayadmit_'+y})

y = 'b18'
b18 = b18.rename(columns={
'Plan Group':'PG_'+y,
'Time Period: Incurred Year':'date_'+y,
'Member Months':'mm_'+y,
'Dollars':'dollars_'+y,
'ER Visits':'er_'+y,
'Admits':'admits_'+y,
'Days Admit':'dayadmit_'+y})

y = 'b19'
b19 = b19.rename(columns={
'Plan Group':'PG_'+y,
'Time Period: Incurred Year':'date_'+y,
'Member Months':'mm_'+y,
'Dollars':'dollars_'+y,
'ER Visits':'er_'+y,
'Admits':'admits_'+y,
'Days Admit':'dayadmit_'+y})

y = 'c18'
c18 = c18.rename(columns={
'Plan Group':'PG_'+y,
'Time Period: Incurred Year':'date_'+y,
'Member Months':'mm_'+y,
'Dollars':'dollars_'+y,
'ER Visits':'er_'+y,
'Admits':'admits_'+y,
'Days Admit':'dayadmit_'+y})

y = 'c19'
c19 = c19.rename(columns={
'Plan Group':'PG_'+y,
'Time Period: Incurred Year':'date_'+y,
'Member Months':'mm_'+y,
'Dollars':'dollars_'+y,
'ER Visits':'er_'+y,
'Admits':'admits_'+y,
'Days Admit':'dayadmit_'+y})
```

```python
comb = a18.merge(a19, how='outer',on='County').merge(b18, how='outer',on='County').merge(b19, how='outer',on='County').merge(c18, how='outer',on='County').merge(c19, how='outer',on='County')
```

```python
fin = ca.merge(comb,how='left',left_on='NAME',right_on='County')
```

```python
fin.to_file('health_final.geojson', driver='GeoJSON')
```

```python
breaks = jp.jenks_breaks((fin['mm_c19'] + fin['mm_c18'] +fin['mm_b19'] + fin['mm_b18']+fin['mm_a19'] + fin['mm_a18'])/6 , nb_class=3)
[ round(elem, -3) for elem in breaks ]
```
